```python
import pandas as pd
import ta
```

```python
data = pd.read_csv('datasets/49_Industry_Portfolios_Daily.csv',skiprows=[0,1,2,3,4,5,6,7,8])
```

```python
#exclude last entry indicating coopyrights
data = data.iloc[0:data.shape[0]-1]

#create date variable
data['Date'] = data.iloc[:,0]
#data['Index'] =  [i for i in range(data.shape[0])]


#drop unnamed column
data = data.drop(['Unnamed: 0'],axis=1)

```

```python
data
```

```python

#select equally weighted
data = data.iloc[25128:data.shape[0]]
data = data.reset_index()
```

```python
data
```

```python
#converts all entries into floats and Date variable into TimeStemp
from datetime import datetime

for col in data.columns:
    if col != 'Date':
        data[col] = [float(str(x).strip()) for x in data[col].values]
    else:
        data[col] = [datetime.strptime(str(x).strip(), '%Y%m%d') for x in data[col].values]

```
